```python
df=pd.read_csv('/content/hotel-reservations-classification-dataset/Hotel Reservations.csv')
df
```

```python
def encoding(x):
  return 1 if x=='Not_Canceled' else 0 
```

```python
df['booking_status']=df['booking_status'].apply(encoding)
```

```python
df
```

```python
df = df.fillna(df.mean())
```

```python
X=df[['no_of_week_nights','lead_time','arrival_month','avg_price_per_room','no_of_special_requests','arrival_year','repeated_guest']]
y=df['booking_status']
X
```

```python
y.value_counts()
```

```python
# from sklearn.preprocessing import LabelEncoder
# le=LabelEncoder()
# y=le.fit_transform(y)
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test=train_test_split(X, y,test_size=0.15, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

clf = RandomForestClassifier(max_depth=2, random_state=0)

clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
acc=accuracy_score(y_test,y_pred)
acc
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    "n_estimators": [10, 20, 30, 40, 50],
    "max_depth": [None, 5, 10, 15, 20],
    "min_samples_split": [2, 4, 6, 8, 10],
    "min_samples_leaf": [1, 2, 4, 6, 8]
}
```

```python
grid_search = GridSearchCV(clf, param_grid, cv=5, scoring="accuracy")
grid_search.fit(X_train, y_train)
```

```python
 grid_search.best_params_
```

```python
best_clf = grid_search.best_estimator_
best_clf.fit(X_train, y_train)
y_pred = best_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
accuracy
```
